```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fx(x):
    return 3*x**2 + 5*x + 2

def dfx(x):
    return 6*x + 5
```

```python
x = np.linspace(-6, 6, 2001)
```

### **Gradient Descent with Fixed Learning Rate**

```python
start_point = np.random.choice(x,1)
localmin = start_point

learning_rate = 0.01
training_epochs = 50 

modelparamFixed = np.zeros((training_epochs,3))

for i in range(training_epochs):
    gradient = dfx(localmin)
    localmin = localmin - learning_rate * gradient

    modelparamFixed[i,0] = localmin
    modelparamFixed[i,1] = gradient
    modelparamFixed[i,2] = learning_rate

plt.plot(x,fx(x))
plt.plot(x,dfx(x))
plt.plot(start_point,fx(start_point),"bo")

plt.axhline(y=0, color='black', linestyle='-')
plt.axvline(x=0, color='black', linestyle='-')

plt.plot(localmin,fx(localmin),"ro")
plt.show()
```

### **G.D. using a gradient-based learning rate**

```python
localmin = np.random.choice(x,1)
initial_value = localmin[:]

learning_rate = 0.01
training_epochs = 50 

modelparamGrad = np.zeros((training_epochs,3))

for i in range(training_epochs):

    gradient = dfx(localmin)

    lr = learning_rate*np.abs(gradient)

    localmin = localmin - lr * gradient

    modelparamGrad[i, 0] = localmin
    modelparamGrad[i, 1] = gradient
    modelparamGrad[i, 2] = lr

```

### **Gradient Descent with Time varying Learning Rates**

```python
localmin = np.random.choice(x,1)
initial_value = localmin[:]

learning_rate = 0.01
training_epochs = 50 

modelparamTime = np.zeros((training_epochs,3))

for i in range(training_epochs):

    gradient = dfx(localmin)

    lr = learning_rate* (1 - ((i+1)/training_epochs))

    localmin = localmin - lr * gradient

    modelparamTime[i, 0] = localmin 
    modelparamTime[i, 1] = gradient
    modelparamTime[i, 2] = lr
```

```python
fig,ax = plt.subplots(1,3,figsize=(15,5))

for i in range(3):
  ax[i].plot(modelparamFixed[:,i],'o-',markerfacecolor='w')
  ax[i].plot(modelparamGrad[:,i],'o-',markerfacecolor='w')
  ax[i].plot(modelparamTime[:,i],'o-',markerfacecolor='w')
  ax[i].set_xlabel('Iteration')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')
ax[2].set_ylabel('Learning rate')
ax[2].legend(['Fixed l.r.','Grad-based l.r.','Time-based l.r.'])

plt.tight_layout()
plt.show()
```
