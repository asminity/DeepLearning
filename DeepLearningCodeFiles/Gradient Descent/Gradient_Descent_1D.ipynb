```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python

def fx(x):
  return 3*x**2 - 3*x + 4

def dx(x):
  return 6*x - 3
```

```python
x = np.linspace(-2,2 ,2001)

plt.plot(x, fx(x))
plt.plot(x, dx(x))
plt.grid()
plt.show()
```

```python
localmin = np.random.choice(x,1)

learning_rate = 0.01
training_epochs = 1000

for i in range(training_epochs):
  grad = dx(localmin)
  localmin = localmin - learning_rate*grad


print(localmin)
print(fx(localmin))
```

```python
plt.plot(x, fx(x))
plt.plot(x, dx(x))
plt.plot(localmin, fx(localmin), 'ro')
plt.plot(localmin, dx(localmin), 'ro')
plt.grid()
plt.show()
```

```python
localmin = np.random.choice(x, 1)

learning_rate = 0.01
training_epochs = 1000
arr = []

for i in range(training_epochs):
    grad = dx(localmin)
    localmin = localmin - learning_rate * grad
    arr.append(localmin[0])

arr = np.array(arr)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(arr, fx(arr), 'ro')
ax[0].set_title("function")


ax[1].plot(dx(arr), 'bo')
ax[1].set_title("derivative")

plt.tight_layout()
plt.show()
```

# **Modify the code so that training ends when some threshold (e.g. 0.1) is reached. Make sure your code is robust for negative derivatives.**

```python
learning_rate = 0.01
training_epochs = 10_000
threshold = 0.1
localmin = np.random.choice(x, 1)
arr = []

for i in range(training_epochs):
  grad = dx(localmin)

  if abs(grad) < threshold:
    break

  localmin = localmin - learning_rate*grad
  arr.append(localmin[0])

arr = np.array(arr)

plt.plot(arr)
```
